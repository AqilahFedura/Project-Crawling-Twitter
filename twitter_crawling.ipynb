```python
import pandas as pd

# Specify the path to your CSV file
file_path = f"tweets-data/{filename}"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path, delimiter=";")

# Initialize an empty 'tags' column
df['tags'] = None

# Fungsi untuk menandai baris berdasarkan kata kunci
def assign_tag(row):
    tags = []
    for keyword, _, _ in keywords_and_date_ranges:
        if keyword.lower() in row['full_text'].lower():
            tags.append(keyword)
    return ', '.join(tags) if tags else None


# Terapkan fungsi ke DataFrame untuk mengisi kolom 'tags'
df['tags'] = df.apply(assign_tag, axis=1)

# Hapus baris yang memiliki 'tags' bernilai None
df = df.dropna(subset=['tags'])

# Reset ulang indeks DataFrame
df = df.reset_index(drop=True)

# Ubah format kolom 'created_at' menjadi "YYYY-MM-DD HH:MM:SS"
df['created_at'] = pd.to_datetime(df['created_at']).dt.strftime('%Y-%m-%d %H:%M:%S')

# Display the DataFrame
#display(df)
```
